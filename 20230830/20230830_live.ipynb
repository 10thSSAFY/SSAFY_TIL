```python
def f(i, N):
    if i == N:
        print(B)
        return
    else:
        B[i] = A[i]
        f(i+1, N)

N = 5
A = [1, 2, 3, 4, 5]
B = [0] * N
f(0, N)
```

```python
def f(i, N, key, arr):  # i 현재 상태, N 목표, key 찾고자 하는 원소
    if i == N:
        return 0  # key가 없는 경우
    elif arr[i] == key:
        return 1
    
    return f(i+1, N, key, arr)

N = 5
A = [1, 2, 3, 4, 5]
key = 10
print(f(0, N, key, A))
```

```python
def f(i, N):
    if i == N:  # 순열 완성
        print(p)
        return
    else:  # card[i]에 들어갈 숫자를 결정
        for j in range(N):
            if used[j] == 0:
                p[i] = card[j]
                used[j] = 1
                f(i+1, N)
                used[j] = 0
                
card = list(map(int, input()))
used = [0]*6  # 이미 사용한 카드인지 표시
p = [0] * 6
f(0, 6)
```

```python
def f(i, N, K):  # N개에서 K개를 고르는 순열
    global cnt
    if i == K:  # 순열 완성: K 개를 모두 고른 경우
        cnt += 1
        print(cnt, p)
        return
    else:  # card[i]에 들어갈 숫자를 결정
        for j in range(N):
            if used[j] == 0:
                p[i] = card[j]
                used[j] = 1
                f(i+1, N, K)
                used[j] = 0
                
card = [1, 2, 3, 4, 5]
N = 5  # N개의 숫자에서
K = 3  # K개를 골라 만드는 수열
used = [0]*N  # 이미 사용한 카드인지 표시
p = [0] * K
cnt = 0
f(0, N, K)
```
