```python
csv_path = 'austin_weather.csv'
df = pd.read_csv(csv_path)
df['Events'] = df['Events'].replace(' ', 'No Events')
df = df[['Date', 'Events']]

data = df.Events.str.split(' , ')
df = pd.DataFrame(data = data)

df = df.explode('Events')
df.astype('str')
df = df['Events'].value_counts()

# print(df.index)
# print(df.values)

plt.bar(df.index, df.values)
plt.grid()

# plt.show()
```
