````python
V = int(input())  #   =   
E = V - 1  # tree    =   - 1
arr = list(map(int, input().split()))
#    
ch1 = [0] * (V+1)
ch2 = [0] * (V+1)
for i in range(E):
    p, c = arr[i*2], arr[i*2+1]
    if ch1[p] == 0:
        ch1[p] = c
    else:
        ch2[p] = c

print(ch1)
print(ch2)
````

````python
# 
def preorder(n):
    if n:  #  
        print(n, end=' ')  # visit(n)
        preorder(ch1[n])  # 
        preorder(ch2[n])  # 

V = int(input())  #   =   
E = V - 1  # tree    =   - 1
arr = list(map(int, input().split()))
#    
ch1 = [0] * (V+1)
ch2 = [0] * (V+1)
for i in range(E):
    p, c = arr[i*2], arr[i*2+1]
    if ch1[p] == 0:
        ch1[p] = c
    else:
        ch2[p] = c

preorder(1)
````

````python
def preorder(n):
    if n:  #  
        print(n, end=' ')  # visit(n)
        preorder(ch1[n])  # 
        preorder(ch2[n])  # 

V = int(input())  #   =   
E = V - 1  # tree    =   - 1
arr = list(map(int, input().split()))
#    
ch1 = [0] * (V+1)
ch2 = [0] * (V+1)
par = [0] * (V+1)
for i in range(E):
    p, c = arr[i*2], arr[i*2+1]
    if ch1[p] == 0:
        ch1[p] = c
    else:
        ch2[p] = c
    par[c] = p  #    

#   
root = 1
while par[root] != 0:
    root += 1
print(root)  # 3
````
