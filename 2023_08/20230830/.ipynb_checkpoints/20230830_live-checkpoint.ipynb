```python
def f(i, N):
    if i == N:
        print(B)
        return
    else:
        B[i] = A[i]
        f(i+1, N)

N = 5
A = [1, 2, 3, 4, 5]
B = [0] * N
f(0, N)
```

```python
def f(i, N, key, arr):  # i  , N , key   
    if i == N:
        return 0  # key  
    elif arr[i] == key:
        return 1
    
    return f(i+1, N, key, arr)

N = 5
A = [1, 2, 3, 4, 5]
key = 10
print(f(0, N, key, A))
```

```python
def f(i, N):
    if i == N:  #  
        print(p)
        return
    else:  # card[i]   
        for j in range(N):
            if used[j] == 0:
                p[i] = card[j]
                used[j] = 1
                f(i+1, N)
                used[j] = 0
                
card = list(map(int, input()))
used = [0]*6  #    
p = [0] * 6
f(0, 6)
```

```python
def f(i, N, K):  # N K  
    global cnt
    if i == K:  #  : K    
        cnt += 1
        print(cnt, p)
        return
    else:  # card[i]   
        for j in range(N):
            if used[j] == 0:
                p[i] = card[j]
                used[j] = 1
                f(i+1, N, K)
                used[j] = 0
                
card = [1, 2, 3, 4, 5]
N = 5  # N 
K = 3  # K   
used = [0]*N  #    
p = [0] * K
cnt = 0
f(0, N, K)
```
