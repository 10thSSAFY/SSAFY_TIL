## 중위순회 문제

```
[입력]
8
1 W 2 3
2 F 4 5
3 R 6 7
4 O 8
5 T
6 A
7 E
8 S
```

````python
def inorder(p, N):
    if p <= N:
        inorder(p*2, N)
        print(tree[p], end='')
        inorder(p*2+1, N)

T = 1
for tc in range(1, T+1):
    N = int(input())
    tree = [0] * (N+1)
    for _ in range(N):
        arr = list(input().split())
        tree[int(arr[0])] = arr[1]
    # 중위순회
    print(f'#{tc}', end=' ')
    inorder(1, N)
````
